# Named entity recognition

```python
# import nltk resources
import nltk
from nltk.book import text1
#import nltk
#nltk.download('words')
#nltk.download('maxent_ne_chunker')

from nltk.tokenize import word_tokenize
from nltk.tokenize import sent_tokenize
```

```python


# make a text string from nltk.book.text1
#text1 = ' '.join(nltk.book.text1)
text1 = "My name is Debasish Sahoo. I study in accepted university I stay in Mayfield Heights. I am from India, Odisha, Cuttack - 753014. My phone number is 4406819126 or (440)-681-9126. My DOB is 19-Apr-1984"
```

```python
# Tokenize the article into sentences: sentences
sentences = sent_tokenize(text1)

# Tokenize each sentence into words: token_sentences
token_sentences = [word_tokenize(sent) for sent in sentences]

# Tag each tokenized sentence into parts of speech: pos_sentences
pos_sentences = [nltk.pos_tag(sent) for sent in token_sentences] 

# Create the named entity chunks: chunked_sentences
chunked_sentences = nltk.ne_chunk_sents(pos_sentences, binary=True)

# dictionary to store NEs
NE = {}

# Test for stems of the tree with 'NE' tags
for sent in chunked_sentences:
    for chunk in sent:
        if hasattr(chunk, "label") and chunk.label() == "NE":
            z = str(chunk) # to make the chunk idexable
            NE.setdefault(z, 0)
            NE[z] += 1
```

```python
# Print out the NE chunks
for z in NE: print(NE[z], "\t", z)
```

```python
for w in sorted(NE, key=NE.get, reverse=True):
    print(NE[w], "\t", w)
```
