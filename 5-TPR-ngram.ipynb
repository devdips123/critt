# Read Study Tables

```python
import pandas as pd
import glob

# path to the TPRDB
TPRDBpath = "/data/critt/tprdb/TPRDB/"

def readTPDDBtables(studies, t):
    df = pd.DataFrame()
    for study in studies:
        for fn in glob.glob(TPRDBpath + study + t):
            df = pd.concat([df, pd.read_csv(fn, sep="\t", dtype=None)], ignore_index=True)
        
    return(df)
```

```python
# read all tables of a study for User P05

study = ["CS19/Tables/"]
STdf = readTPDDBtables(study, "P05*.st")
TTdf = readTPDDBtables(study, "P05*.tt")
```
