In our previous semester's class, all the students ( grad and under-grad) were given a set of tasks(Copy, Paraphrase, Summerize) a set of predefined studies to generate target texts. These tasks were performed in our lab with computers configured with EyeTracker (SMI250) and KeyStroke Logger (Translog-II). These data was further processed to align source token to target token by the users and then finally stored to the TPR(Translation Process Research) Database.

*EyeTracker* captures the user's gaze time, the duration user focuces on each word on source and target token etc.<br/>
*Translog* captures all the information regarding the key input through the keyboard like Num of Insertions, Deletions, Time taken to input each word token, pauses, etc. 

We are trying to find the effect of the existence of orthographic and semantic neighbors on the word production time of the word-tokens in our empirical data stored in TPR DB.

According to the popular research papers BIA, BIA+, Multilink, when an user is producing/translating text, similar words get activated in the mind of the user. The similarity of the words can be categorized into 3 types - Orthographic(similar looking), Semantic(similar meaning), Phonetic(similar sounding). Such activation in the user's mind may affect the word-production time.

We in our experiment, will use certain metrics like Levenshtein's Distance and Word2Vec to generate the similar words for each word token in our experimental data, and then try to establish a relationship between these words and the word-production time.

#### Orthographic Neigbours using Levinshtein's Distance

According to BIA, Orthographic Neighbours of a token are those words which differ just by 1 character. We use Levinshtein's Distance to find the orthographically similar-word list.


We are using the BNC (British National Corpus) to use as a base repository of word-tokens to generate the orthographic similar words.

We perform the following operations on BNC

- Sentence tokenize and then word tokenize to get all word-tokens from the BNC
- Group same words and create a dictionary of each distict word and its frequency
- remove the numeric tokens, alpha-numeric, tokens with special characters etc. 
- Ignore words with freq < 10 since these might be typing errors and wouldnt add value to our experiment.
- Store the resulting tokens as a dictionary with word-length as key and value as (dictionary of word and its frequency). 
  Such a format would be helpful to us to find Levenshtein's Distance
  
The Levenshtein Distance(LD) is a string metric for measuring the difference between two sequences.
Informally, the LD between two words is the minimum
number of single-character edits (insertions, deletions or
substitutions) required to change one word into the other
