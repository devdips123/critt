# Relation beween two variables
- Correlations 
- Regression

```r
# Correlate number of deletions (del) and production time (Dur)

RUC_P1st_dur <- RUC_P1st[RUC_P1st$Dur > 0 ,]$Dur

#correlation
cor(RUC_P1st$Dur, RUC_P1st$Del, method="pearson")

#Linear regression
LM_Dur_Del <- lm(Dur ~ Del, data=RUC_P1st)
summary(LM_Dur_Del)

# create scatterplot
plot(Dur~Del, data=RUC_P1st)

#insert regression line
abline(LM_Dur_Del, col="red")
```

```r
RUC_P1st_dur <- RUC_P1st[RUC_P1st$Dur > 0 ,]
RUC_P1st_dur$DurLog <- log(RUC_P1st_dur$Dur)

#correlation
cor(RUC_P1st_dur$DurLog, RUC_P1st_dur$Del, method="pearson")

#Linear regression
LM_Dur_Del <- lm(DurLog ~ Del, data=RUC_P1st_dur)
summary(LM_Dur_Del)

# create scatterplot
plot(DurLog ~ Del, data=RUC_P1st_dur)

#insert regression line
abline(LM_Dur_Del, col="red")
```

```r
#install.packages("effects")
library("effects")
	
#plot the linear model
plot(allEffects(LM_Dur_Del))
```

# Read all session files of a study and look at some values

- read the st table
- inspect the dataframe
- some basic statistical distributions

```r
library(readr)

# read all tables of type 'tab' in all studies under path into a dataframe
readCRITTables <- function(studies, tab, path="/data/critt/tprdb/TPRDB") {
  D <- {}
  for (study in studies) {
    files = list.files(paste(path, study, sep="/"), full.names=T, pattern=tab)
    for (i in files) {
      D <- rbind(D, read_delim(i, "\t", escape_double = FALSE, col_types = cols(Task = col_character()), na = "---", trim_ws = TRUE))
  } }
  return(D)
}
```

```r
DFst <- readCRITTables(c("BML12/Tables/"), "*.st")
```

```r
# repeat analysis of previous section 
```

```r
# size of the dataframe rows, columns
dim(DFst)

# number of rows
nrow(DFst)

# number of clolumns
ncol(DFst)
```

```r
hist(DFst[DFst$Dur > 0, ]$Dur)
hist(log(DFst[DFst$Dur > 0, ]$Dur))
```

```r
plot(density(log(DFst[DFst$Dur > 0, ]$Dur)))
```

```r
#Correlate between Dur and Del
dur = DFst[DFst$Dur > 0, ]$Dur
del = DFst[DFst$Dur > 0, ]$Del
paste("Pearson", cor(dur,del, method='pearson'))
paste("Spearman",cor(dur,del, method='spearman'))
```

```r

#DFst$Dur
res_durDEL = DFst[(DFst$Dur > 0) & (DFst$Del > 0), ]
res_durDEL$logDur = log(res_durDEL$Dur)
res_durDEL$logDel = log(res_durDEL$Del)
```

```r
head(res_durDEL[,c('logDur','logDel')])
```

```r
####1. Correlate number of deletions (del) and production time (Dur)
#create scatterplot
plot(logDur~logDel, data=res_durDEL)
#correlation
cor(res_durDEL$Dur, res_durDEL$Del, method="pearson")
#Linear regression
LM_Dur_Del <- lm(logDur~logDel, data=res_durDEL)
summary(LM_Dur_Del)
#insert regression line
abline(LM_Dur_Del, col="red")
```
